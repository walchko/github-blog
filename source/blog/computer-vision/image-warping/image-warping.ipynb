```python
import cv2         # opencv itself
import numpy as np # matrix manipulations

from matplotlib import pyplot as plt 
```

```python
img = cv2.imread('cb.jpg')
img = cv2.cvtColor(img, cv2.COLOR_BGR2RGB)
plt.imshow(img)
plt.title('{}'.format(img.shape));
```

```python
flags = 0
marker_size = (7,10)
# imgpoints = []

ret, corners = cv2.findChessboardCorners(img, marker_size, flags=flags)
# imgpoints.append(corners.reshape(-1, 2))
```

```python
# make a grid of points
objp = np.zeros((np.prod(marker_size), 3), np.float32)
objp[:, :2] = np.indices(marker_size).T.reshape(-1, 2)
```

```python
h, stat = cv2.findHomography(objp, corners)
```

```python
h
```

```python
warp_img = cv2.warpPerspective(img, h, (img.shape[1], img.shape[0]))
```

```python
plt.imshow(warp_img)
```
