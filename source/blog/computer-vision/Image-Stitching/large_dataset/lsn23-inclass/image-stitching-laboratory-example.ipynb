Let's create a stitching image of our Maker Lab.

```python
%matplotlib inline
```

```python
from __future__ import print_function
from __future__ import division
import matplotlib.pyplot as plt
import numpy as np
import cv2
from math import sqrt
import pylab                                   
pylab.rcParams['figure.figsize'] = (10.0, 8.0)
```

```python
fname = [
    'new-pics/IMG_4150.jpg',
    'new-pics/IMG_4151.jpg',
    'new-pics/IMG_4152.jpg',
    'new-pics/IMG_4153.jpg',
    'new-pics/IMG_4154.jpg',
    'new-pics/IMG_4155.jpg',
    'new-pics/IMG_4156.jpg',
    'new-pics/IMG_4157.jpg',
    'new-pics/IMG_4158.jpg'
]

imgs2 = []
for f in fname:
    image = cv2.imread(f)
    image = cv2.cvtColor(image,cv2.COLOR_BGR2RGB)
    imgs2.append(image)

# display the individual images
plt.figure(figsize=(9,6))  # width, height
for i, img in enumerate(imgs2):
    plt.subplot(3,3,i+1)
    plt.imshow(img)
    h,w,_ = img.shape
    plt.title('{}x{}'.format(w,h))
    plt.xticks(())
    plt.yticks(())
```

```python
stitcher = cv2.createStitcher(True)  # use the GPU
err, result = stitcher.stitch(imgs2)
if err:
    print("crap!")
    
# save a high res copy to look at
ans = cv2.cvtColor(result,cv2.COLOR_RGB2BGR)
cv2.imwrite('lab.png', ans)

# display
plt.imshow(result);
h,w,_ = ans.shape
plt.title('{}x{}'.format(w,h));
```
