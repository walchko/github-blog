Given some data, how can we use `pandas` to display the data in a meaningful way?

```python
import pandas as pd
import numpy as np
```

```python
# animals = {
#     'falcon': {'legs': 2, 'wings': 2, 'seen': 10},
#     'robin': {'legs': 2, 'wings': 2, 'seen': 8},
#     'bluejay': {'legs': 2, 'wings': 2, 'seen': 4},
#     'dog': {'legs': 4, 'wings': 0, 'seen': 22},
#     'cat': {'legs': 4, 'wings': 0, 'seen': 12},
#     'squirl': {'legs': 4, 'wings': 0, 'seen': 28},
#     'spider': {'legs': 8, 'wings': 0, 'seen': 11},
#     'ant': {'legs': 6, 'wings': 0, 'seen': 28},
#     'fish': {'legs': 0, 'wings': 0, 'seen': 8},
#     'horse': {'legs': 4, 'wings': 0, 'seen': 5},
#     'unicorn': {'legs': 4, 'wings': 0, 'seen': 0},
#     'pegasus': {'legs': 4, 'wings': 2, 'seen': 0}
# }
animals = {
    0:  {'name': 'falcon', 'legs': 2, 'wings': 2, 'seen': 10},
    1:  {'name': 'robin',  'legs': 2, 'wings': 2, 'seen': 8},
    2:  {'name': 'bluejay','legs': 2, 'wings': 2, 'seen': 4},
    3:  {'name': 'fish',   'legs': 0, 'wings': 0, 'seen': 8},
    4:  {'name': 'horse',  'legs': 4, 'wings': 0, 'seen': 5},
    5:  {'name': 'dog',    'legs': 4, 'wings': 0, 'seen': 22},
    6:  {'name': 'cat',    'legs': 4, 'wings': 0, 'seen': 12},
    7:  {'name': 'squirl', 'legs': 4, 'wings': 0, 'seen': 28},
    8:  {'name': 'spider', 'legs': 8, 'wings': 0, 'seen': 11},
    9:  {'name': 'ant',    'legs': 6, 'wings': 0, 'seen': 28},
    10: {'name': 'unicorn','legs': 4, 'wings': 0, 'seen': 0},
    11: {'name': 'pegasus','legs': 4, 'wings': 2, 'seen': 0}
}

df = pd.DataFrame(animals)
df = df.transpose()
df.head()
```

```python
df.plot.pie(y="legs", labels=df["name"], legend=False);
```

```python
df.plot.scatter(x='wings', y='legs', c="seen", colormap="viridis", grid=True, title="stuff");
```

```python
df.plot.bar(grid=True, title="Simple Bar Chart", x="name");
```

```python
df.plot.bar(grid=True, title="Stacked Bar Chart", x="name", stacked=True);
```

```python
df.plot.barh(grid=True, title="Stacked Bar Chart on Its Side", x="name", stacked=True);
```

```python
df.plot.hist(bins=20, grid=True, title="Histogram");
```

```python
df.plot.hist(bins=10, grid=True, title="Histogram", subplots=True);
```

```python
df.plot(kind='density', subplots=True, layout=(2,2), sharex=False, grid=True);
```

```python
df.plot.box(grid=True);
```

```python
df.plot.box(grid=True, subplots=True, sharex=False, sharey=False);
```

# Matplotlib

You can do most of the same things (and more) with Matplotlib. Here are a couple examples:

```python
from matplotlib import pyplot as plt
%matplotlib inline
from mpl_toolkits.mplot3d import Axes3D
```

```python
for col in ['wings', 'legs']:
    plt.hist(df[col], alpha=0.5, label=col)
plt.legend()
plt.grid(True)
```

```python
# let's manipulate the data and create a new column
df["metric"] = (df.legs - df.wings)/2

# now let's create a new 3d plot and fill it with data
p = plt.figure().gca(projection="3d")
p.scatter(df["legs"], df["wings"], df["metric"], depthshade=False)
p.set_xlabel("legs")
p.set_ylabel("wings")
p.set_zlabel("metric")
p.set_title("My Awesome 3d Plot");
```

```python
df.plot.scatter(x="metric", y="legs", grid=True);
```
